## Word Embedding using Embedding Layer in Keras

```python
import keras
```

```python
keras.__version__
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##sentence
snt=[ 'the glass of milk',
     'the glass of juice',
    'the cup of tea',
    'i am good boy',
    'I am good devoloper',
    'understand the meaning of word',
    'your videos are good',]
```

```python
snt
```

```python
## vocabulary size
voc_size=10000
```

### One Hot Representation

```python
onehot_repr = [one_hot(words,voc_size) for words in snt]
print(onehot_repr)  ##index from dictionary
```

### Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences  ## in every sentence same number of words(same size)
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(onehot_repr, padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## dimension
dim=10
```

```python
model = Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
